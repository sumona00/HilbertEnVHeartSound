```python
import matplotlib.pyplot as plt
from matplotlib.backend_bases import RendererBase
from scipy import signal
from scipy.io import wavfile
#import soundfile as sf
import os
import numpy as np
from PIL import Image
from scipy.fftpack import fft
from pathlib import Path
import glob
import numpy as np
import arff
import soundfile as sf
%matplotlib inline
```

```python
def log_specgram(audio, sample_rate, window_size=20,
                 step_size=10, eps=1e-10):
    nperseg = int(round(window_size * sample_rate / 1e3))
    noverlap = int(round(step_size * sample_rate / 1e3))
    freqs, times, spec = signal.spectrogram(audio,
                                    fs=sample_rate,
                                    window='hann',
                                    nperseg=nperseg,
                                    noverlap=noverlap,
                                    detrend=False)
    return freqs, times, np.log(spec.T.astype(np.float32) + eps)
```

```python
data_dir = Path('/home/sumona/Env/MVP_clean')
#Normal=1 #MS=2 #MS=3 #AS=4 #MVP=5
sample_audio = []
labels=[]
f_mat=np.empty([1,384])
x_wav=[]
for filename in glob.glob(os.path.join(data_dir,'*.wav')):
    q=os.path.join(data_dir,filename)
    print(q)
    # Read wav with default
    x, Fs = sf.read(q)
    s=q+'.arff'
    print(s)
    dataset=arff.load(open(s, 'r'))
    data=np.array(dataset['data'])
    attributes = dataset['attributes']
    a=data
    a=np.delete(a,0,1)
    a=np.delete(a,384,1)
    f_mat=np.append(f_mat,a,axis=0)
    sample_audio.append(q)
    labels.append(5)
    x_wav.append(x)
f_mat=np.delete(f_mat,0,0)
labels=np.asarray(labels)
x_wav=np.asarray(x_wav)
labels.shape
print(f_mat)
f_mat.shape
print(sample_audio)
sample_audio=np.asarray(sample_audio)
sample_audio.shape
fig_loc=[]
spect_res=[]
for r in sample_audio:
    sample_rate, samples = wavfile.read(r)
    freqs, times, spectrogram = log_specgram(samples, sample_rate)
    spect_res.append(spectrogram)
    fig = plt.figure(figsize=(14, 8))
    plt.imshow(spectrogram.T, aspect='auto', origin='lower', 
           extent=[times.min(), times.max(), freqs.min(), freqs.max()])
    plt.savefig(r + ".jpg")
    fig_loc.append(r +".jpg")

spect_res=np.asarray(spect_res) 
spect_res.shape
print(fig_loc)
fig_loc=np.asarray(fig_loc)
fig_loc.shape
import tensorflow as tf 
from keras.applications.resnet50 import ResNet50
from keras.layers import Flatten, Input
from keras.models import Model
from keras.preprocessing import image
from keras.applications.imagenet_utils import preprocess_input
import numpy as np
from io import BytesIO
import time
import cv2
start = time.time()
model = ResNet50(weights='imagenet', pooling=max, include_top = False) 
model.summary()
im=[]
spect_features=[]
for file in fig_loc:
    f = cv2.imread(file)
    im.append(f)
    img = image.load_img(file, target_size=(224, 224))
    x = image.img_to_array(img) 
    x = np.expand_dims(x, axis=0) 
    x = preprocess_input(x) 
    features = model.predict(x)
    features_reduce = features.squeeze()
    features_flatten = features_reduce.reshape((features.shape[0], 7 * 7 * 2048))
    spect_features.append(features_flatten)
spect_features=np.asarray(spect_features)
spect_features=spect_features.reshape(200,100352)
spect_features.shape
features=np.concatenate((f_mat,spect_features), axis=1)
features=np.concatenate((features,x_wav), axis=1)
from numpy import savetxt
savetxt('features_MVP_clean.csv',features)
savetxt('labels_MVP_clean.csv',labels)
from numpy import loadtxt
data = loadtxt('features_MVP_clean.csv')
data.shape
labels=loadtxt('labels_MVP_clean.csv')
print(labels)
```
