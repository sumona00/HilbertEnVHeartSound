```python
from sklearn.metrics import precision_recall_fscore_support

def plot_multiclass_roc(model, X_test, y_test, n_classes, figsize=(10, 5)):
    y_score = model.predict(X_test)
    
    # y_test=y_test.ravel()
    # structures
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    # calculate dummies once
    y_test_dummies = pd.get_dummies(y_test, drop_first=False).values
    for i in range(n_classes):
        fpr[i], tpr[i], threshold = metrics.roc_curve(np.array(pd.get_dummies(y_test))[:, i], np.array(pd.get_dummies(y_pred))[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])
      
    # roc for each class
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('Receiver operating characteristic example')
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f) for label %i' % (roc_auc[i], i))
    ax.legend(loc="best")
    ax.grid(alpha=.4)
    sns.despine()
    plt.show()
    
def accuracy(confusion_matrix):
    diagonal_sum = confusion_matrix.trace()
    sum_of_all_elements = confusion_matrix.sum()
    return diagonal_sum / sum_of_all_elements 
def precision(label, confusion_matrix):
    col = confusion_matrix[:, label]
    return confusion_matrix[label, label] / col.sum()
    
def recall(label, confusion_matrix):
    row = confusion_matrix[label, :]
    return confusion_matrix[label, label] / row.sum()

def precision_macro_average(confusion_matrix):
    rows, columns = confusion_matrix.shape
    sum_of_precisions = 0
    for label in range(rows):
        sum_of_precisions += precision(label, confusion_matrix)
    return sum_of_precisions / rows

def recall_macro_average(confusion_matrix):
    rows, columns = confusion_matrix.shape
    sum_of_recalls = 0
    for label in range(columns):
        sum_of_recalls += recall(label, confusion_matrix)
    return sum_of_recalls / columns

def run_test(y_test, y_pred):
    print(accuracy_score(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    print("=== Classification Report ===")
    print(classification_report(y_test, y_pred))
    print('\n')
    skplt.metrics.plot_confusion_matrix(y_test,y_pred, normalize=True)
    plt.show()
    cm1 = metrics.confusion_matrix(y_test,y_pred)

    FP1 = cm1.sum(axis=0) - np.diag(cm1)
    print(FP1)
    FN1 = cm1.sum(axis=1) - np.diag(cm1)
    print(FN1)
    TP1 = np.diag(cm1)
    print(TP1)
    TN1 = cm1.sum() - (FP1 + FN1 + TP1)
    print(TN1)

    # Sensitivity, hit rate, recall, or true positive rate
    SEN1 = np.divide(TP1,np.add(TP1,FN1))
    print(SEN1)
    # Specificity or true negative rate
    SPE1 =np.divide(TN1,np.add(TN1,FP1))
    print(SPE1)
    
    [precision, recall, f1, support]=precision_recall_fscore_support(y_test, y_pred, average='weighted')
   
    print("{:.2f}".format(accuracy_score(y_test, y_pred)*100), "{:.2f}".format(recall*100), \
                           "{:.2f}".format(np.mean(SPE1)*100), "{:.2f}".format(precision*100), \
                           "{:.2f}".format(f1*100))
    
    plot_multiclass_roc(model, X_test, y_test, n_classes=5, figsize=(10, 5))
```
